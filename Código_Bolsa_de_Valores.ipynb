```python
import numpy as np
import yfinance as yf
```

```python
def pegar_retornos(nomeTicker):
  tickerData = yf.Ticker("AAPL")
#SBUX

  tickerDf = tickerData.history(period='1y',interval='1d')
  tickerClose = tickerDf.Close
  daily_returns_nan = tickerClose.pct_change() #pega os retornos
  daily_returns = daily_returns_nan.dropna()
  return daily_returns
```

```python
ap = pegar_retornos("AAPL").values #<- Elimina as datas
msf = pegar_retornos("MSFT").values #<- Elimina as datas
print(ap)
print(msf)
```

```python
cov = np.cov(ap,msf)# <- Isso Retorna a Matrix de Covariancia
```

```python
cov[0][0] + cov[1][1]
cima = 2*cov[0][0]
baixo = 2*(cov[0][0] + cov[1][1])
xb = cima/baixo
xb
```

```python
xa = 1-xb
xa
```

```python
cov[0][0]*xa**2 + cov[0][1] *xb**2
```

```python
ap.mean()*xa + msf.mean()*xb
```
